```python
import pandas as pd
import numpy as np
# Libraries to help with data visualization
import matplotlib.pyplot as plt

import seaborn as  sns
```

```python
customer_read = pd.read_csv('1.csv', sep = ",")
customer_data = pd.DataFrame(customer_read)
```

```python
customer_data.head()
```

```python
customer_data = customer_data.drop(columns='Unnamed: 0')
```

```python
data = customer_data.drop(columns='AcceptedCmp5')
target = customer_data['AcceptedCmp5']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target, test_size = 0.2)
print("X_train: ", X_train.shape)
print("X_test: ", X_test.shape)
print("y_train: ", y_train.shape)
print("y_test: ", y_test.shape)
```

```python
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier(random_state=0)
clf.fit(X_train, y_train)
y_pred_train = clf.predict(X_train)
y_pred_test = clf.predict(X_test)
acc_train = accuracy_score(y_train, y_pred_train)
acc_test = accuracy_score(y_test, y_pred_test)
print("Accuracy score of training data: ", acc_train)
print("Accuracy score of test data: ", acc_test)
```

```python
from sklearn import tree

plt.figure(figsize = (18, 16))
tree.plot_tree(clf)
plt.show()
```
